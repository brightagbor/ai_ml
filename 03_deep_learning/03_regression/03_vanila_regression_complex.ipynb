```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import random
```

```python
# Set random seeds for reproducibility
seed = 42
tf.random.set_seed(seed)
np.random.seed(seed)
random.seed(seed)
```

```python
## Loading the data
dataframe = pd.read_csv("Mobile-Price-Prediction-cleaned_data.csv")
dataframe.head()
```

```python
dataframe.info()
```

```python
dataframe.describe().T
```

```python
# Separating predictors from targets
X = dataframe.drop("Price", axis=1)
y = dataframe["Price"]
```

```python
## Shape of X and y
X.shape, y.shape
```

```python
# Split X and y into train and val sets
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
## Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
```

```python
# For efficiency, convert data to tensors or as a tf_.data.Dataset
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train)).batch(128)
val_dataset = tf.data.Dataset.from_tensor_slices((X_val, y_val)).batch(128)
```

```python
train_dataset
```

```python
for batch in train_dataset:
    print(len(batch[0]))
```

```python
type(train_dataset)
```

```python
input_shape = X_train.shape[1]
input_shape
```

```python
# Define Model using the Sequential API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation="relu", input_shape=(input_shape, )),
    tf.keras.layers.Dense(32, activation='relu'),
    # tf.keras.layers.Dense(64, activation='relu'),
    # tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1) # regression output
])
```

```python
model.summary()
```

```python
## Define Optimizer, Loss and Metrics
optimizer = tf.keras.optimizers.Adam(1e-3)
loss_fn = tf.keras.losses.MeanSquaredError()
optimizer
```

```python
# Configure metrics
train_loss_metric = tf.keras.metrics.Mean(name="train_loss")
val_loss_metric = tf.keras.metrics.Mean(name="val_loss")
train_loss_metric
```

```python
# Training Step Function

@tf.function
def train_step(x, y):
    with tf.GradientTape() as tape:
        y_pred = model(x, training=True)
        loss_value = loss_fn(y, y_pred)
        loss_value += sum(model.losses)
    grads = tape.gradient(loss_value, model.trainable_variables)
    optimizer.apply_gradients(zip(grads, model.trainable_variables))
    train_loss_metric.update_state(loss_value)
```

```python
# Validation Step unction

@tf.function
def val_step(x, y):
    y_pred = model(x, training=False)
    v_loss = loss_fn(y, y_pred)
    val_loss_metric.update_state(v_loss)
```

```python
# Training Loop
EPOCHS = 1000

# Containers for history
history = {
    "train_loss": [],
    "val_loss": [],
}

for epoch in range(EPOCHS):

    # Reset metrics at the start of each epoch
    train_loss_metric.reset_states()
    val_loss_metric.reset_states()

    # Training 
    for x_batch, y_batch in train_dataset:
        train_step(x_batch, y_batch)

    # Validation
    for x_batch, y_batch in val_dataset:
        val_step(x_batch, y_batch)

    # Store results
    train_loss = train_loss_metric.result().numpy()
    val_loss = val_loss_metric.result().numpy()


    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)

    if (epoch+1)%100 == 0:
        print(f"Epoch: {epoch+1} Train Loss: {train_loss_metric.result():.4f} Val Loss: {val_loss_metric.result():.4f}")
```

```python
## Evaluation after training
y_pred = model.predict(X_val)

# Calculate MAE or MSE manually
mse = tf.keras.metrics.mean_squared_error(y_val, y_pred).numpy().mean()
mae = tf.keras.metrics.mean_absolute_error(y_val, y_pred).numpy().mean()

print("Test MSE:", mse)
print("Test MAE:", mae)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_val, y_pred)
score
```
